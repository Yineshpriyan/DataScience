```python
# Grid Plots -
  # It hleps to visualize relationships between multiple variables in one view.
  # Here we can do customizations as we need. That means we can plot different grpahs in a single view

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
# Pair Grid
  # It will create empty grids for pair of all  numerical columns in the dataset.

sns.PairGrid(iris)
```

```python
a = sns.PairGrid(iris)
a.map(plt.scatter)
```

```python
a = sns.PairGrid(iris)
a.map_diag(plt.hist)
a.map_upper(sns.kdeplot)
a.map_lower(plt.scatter)
```

```python
# Facet Grid -
 # In Facet Grid, we can do deep selection/ FIlteration. That means we can add more features in visualization.
```

```python
tip = sns.load_dataset('tips')
tip.head()
```

```python
g = sns.FacetGrid(tip, col='day', row='smoker')
g.map(plt.hist,'total_bill')
```

```python
g = sns.FacetGrid(tip, col='day', row='smoker')
g.map(plt.scatter, 'total_bill','tip')
```

```python
g = sns.FacetGrid(tip, col='day', row='smoker', hue='sex')
g.map(plt.scatter, 'total_bill','tip')
```

```python
g = sns.FacetGrid(tip, col='day', row='smoker', hue='sex')
g.map(plt.scatter, 'total_bill','tip').add_legend()
```

```python
# Joint Grid

g = sns.JointGrid(x='total_bill', y='tip', data=tip)
g.plot(sns.regplot, sns.kdeplot)
```

```python
g = sns.JointGrid(x='total_bill', y='tip', data=tip)
g.plot(plt.plot, sns.kdeplot)
```
