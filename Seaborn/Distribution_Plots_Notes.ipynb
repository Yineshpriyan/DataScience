# Distribution Plot

## distplot / histplot
---------------------

It will works for uni-variant (single variable).

- **distplot** is replaced by **histplot**
- Jointplot
- scatter
- reg
- resid
- kde
- hex

```python
import seaborn as sns
#import matplotlib.pyplot as plt
```

```python
 df = sns.load_dataset('tips')
```

```python
df
```

```python
# distplot

sns.distplot(df['total_bill'])
```

```python
# histplot

sns.histplot(df['total_bill'])
```

```python
# Kernal Density Estimate:
    # Kde estimates the Probility Density function for values. That means shows in probability percentage instaed of showing real values.

sns.histplot(df['total_bill'], kde=True)
```

```python
# Custom Bins

sns.histplot(df['total_bill'], kde=True, bins=30)
```

## Joint Plot

for bivariate (two variables : x, y)

```python
# Scatter

sns.jointplot(x = 'total_bill', y='tip', data=df)
```

```python
# hex
sns.jointplot(x='total_bill', y='tip', data=df, kind='hex', height=5)
```

```python
# reg

sns.jointplot(x='total_bill', y='tip', data=df, kind='reg', height=5)
```

```python
# pairplot
    #is used to create an individual grpah for all the pairs of numerical columns
sns.pairplot(df)
```

```python
# hue
sns.pairplot(df, hue='sex')
```

```python
# palette
sns.pairplot(df, hue='sex', palette='coolwarm')
```

```python
# rugplot
  # univariate
  # this is the building block fro kde plt

sns.rugplot(df['total_bill'])
```

```python
#kde plot
  # This is Keranl Density Estimation plot

sns.kdeplot(df['total_bill'])
sns.rugplot(df['total_bill'])
```

```python
sns.rugplot(df['tip'])
sns.kdeplot(df['tip'])
```
