```python
import seaborn as sns
import pandas as pd
```

```python
t = sns.load_dataset('tips')
t.head()
```

```python
pairgrid - it will create te structure of pairplot, so we can customise the structure
facetgrid - we can add more features in this graph (row, col)
jointgrid - it will create the structture of jointplot, so we can cutomise it.
```

```python
#pairgrid
pg = sns.PairGrid(t)
pg.map(sns.scatterplot)
```

```python
pg = sns.PairGrid(t)
pg.map_diag(sns.histplot)
pg.map_upper(sns.kdeplot)
pg.map_lower(sns.scatterplot)
```

```python
f = sns.FacetGrid(t, row='smoker', col='time')
f.map(sns.countplot, 'total_bill')
```

```python
f = sns.FacetGrid(t, row='smoker', col='time', hue='sex')
f.map(sns.histplot, 'total_bill','tip').add_legend()
```

```python
jg = sns.JointGrid(x='total_bill', y='tip', data=t)
jg.plot(sns.regplot,sns.kdeplot)
```
