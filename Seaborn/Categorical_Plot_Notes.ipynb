Categorical Plot
---------------

- factorplot / catplot
- boxplot
- violinplot
- stripplot
- swarmplot
- barplot
- countplot


```python
import seaborn as sns
import numpy as np
```

```python
df = sns.load_dataset('tips')
df
```

```python
# barplot
  # It is used to check the data balance immediately while check the data balance.
  # default estimator is mean of y axis.
sns.barplot(x='sex' , y='total_bill' , data=df , palette='coolwarm')
```

```python

sns.barplot(x='sex', y='total_bill', data=df, estimator=np.std)
sns.barplot(x='sex', y='total_bill', data=df, palette='coolwarm', estimator=np.std)
```

```python
# countplot - it will show the count values of each category

sns.countplot(x='sex', data=df, palette='coolwarm')
```

```python
# box plot - it is useful to find the outliers in data easily.

sns.boxplot(x='day', y='total_bill', data=df, palette='coolwarm')
```

```python
sns.boxplot(data=df)
```

```python
sns.boxplot(data=df, orient='h')
```

```python
sns.boxplot(x='day', y='total_bill', data=df, hue='sex')
```

```python
# violin plot - It is same as box plot. but here additionally we can see the distribution visualy

sns.violinplot(x='day', y='total_bill', data=df, palette='rainbow')

```

```python
sns.violinplot(x='day', y='total_bill', data=df, hue='sex', palette='rainbow')
```

```python
sns.violinplot(x='day', y='total_bill', data=df, hue='sex', palette='rainbow', split=True)
```

```python
# Stripplot

sns.stripplot(x='day', y='total_bill', data=df, palette='rainbow')
```

```python
sns.stripplot(x='day', y='total_bill', data=df, palette='rainbow', jitter=False)
```

```python
sns.stripplot(x='day', y='total_bill', data=df, palette='rainbow', hue='sex')
```

```python
sns.stripplot(x='day', y='total_bill', data=df, palette='rainbow', jitter=False, hue='sex')
```

```python
sns.stripplot(x='day', y='total_bill', data=df, palette='rainbow', hue='sex', dodge=True)
```

```python
# Swarmplot - same as stripplot but data points won't ovelap in graph.

sns.swarmplot(x='day', y='total_bill', data=df, palette='rainbow')
```

```python
sns.swarmplot(x='day', y='total_bill', data=df, palette='rainbow', hue='sex')
```

```python
sns.swarmplot(x='day', y='total_bill', data=df, palette='rainbow', hue='sex', dodge=True)
```

```python
sns.violinplot(x='day', y='total_bill', data=df, palette='rainbow')
sns.swarmplot(x='day' ,y='total_bill', data=df)
```

```python
sns.violinplot(x='day', y='total_bill', data=df, palette='rainbow', hue='sex', split=True)
sns.swarmplot(x='day' ,y='total_bill', data=df , hue='sex', dodge=True)
```

```python
# Catplot

sns.catplot(x='day' , y='total_bill', data=df, palette='rainbow')
```

```python
sns.catplot(x='day', y='total_bill', data=df, palette='rainbow', kind='bar')
```

```python
sns.catplot(x='day', y='total_bill', data=df, palette='rainbow', kind='swarm')
```
