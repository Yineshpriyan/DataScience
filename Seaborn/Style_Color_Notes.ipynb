```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# set_style - set the backgound of chart (white, dark, whitegrid, darkgrid, ticks)

sns.set_style('white')
sns.countplot(x='day', data=tips)
```

```python
sns.set_style('dark')
sns.countplot(x='day', data=tips)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='day', data=tips)
```

```python
sns.set_style('darkgrid')
sns.countplot(x='day', data=tips)
```

```python
sns.set_style('ticks')
sns.countplot(x='day', data=tips)
```

```python
# despine - add and remove the border line of chart
  # sns.despine() - remove top and right lines.
  # sns.despine(left=True) - remove top, righ and left.
  # sns.despine(right=False) - remove only top.

sns.set_style('ticks')
sns.countplot(x='sex', data=tips)
sns.despine()
```

```python
sns.set_style('ticks')
sns.countplot(x='sex', data=tips)
sns.despine(left=True)
```

```python
sns.set_style('ticks')
sns.countplot(x='sex', data=tips)
sns.despine(left=True, bottom=True)
```

```python
sns.set_style('ticks')
sns.countplot(x='sex', data=tips)
sns.despine(right=False)
```

```python
# figure - we can mention the size of the chart frame/ figure

fig= plt.figure(figsize=(12,6))
sns.set_style('ticks')
sns.countplot(x='sex', data=tips)
sns.despine()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
# aspect, height

sns.lmplot(x='total_bill', y='tip', aspect=0.6, height=5, data=tips)
```

```python
# Scale and Context

sns.set_context('poster')
sns.countplot(x='sex', data=tips)
```

```python
sns.set_context('poster', font_scale=4)
sns.countplot(x='sex', data=tips)
```

```python
sns.set_context('paper')
sns.countplot(x='sex', data=tips)
```

```python
sns.set_context('notebook')
sns.countplot(x='sex', data=tips)
```
