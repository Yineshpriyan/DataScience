```python
# Heatmap - before star working on the dataset we have to check the null values.
  #for that purpose to check that is there any null values or not, we can use heatmap to get the visual
```

```python
import seaborn as sns
import pandas as pd
```

```python
fl = sns.load_dataset('flights')
fl.head()
```

```python
fl['month'] = pd.Categorical(fl['month']).codes
fl.head()
```

```python
fl.corr()
```

```python
sns.heatmap(fl.corr())
```

```python
sns.heatmap(fl.corr(), annot=True, linecolor='white', linewidths=1)
```

```python
# pivot tabe - create a new table from an existing table only with specific columns we need
fl2 = fl.pivot_table(values='passengers', index='month', columns='year')
fl2
```

```python
sns.heatmap(fl2, linecolor='white', linewidths='1')
```
