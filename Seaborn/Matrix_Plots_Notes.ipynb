```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
fl = sns.load_dataset('flights')
fl.head()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
fl['month'] = pd.Categorical(fl['month']).codes
fl
```

```python
fl2=fl.corr()
fl2
```

```python
# Heatmap:
# it uses color density show the data avalue.
# we can use heatmap to find the empty columns and null values in a dataset easily before start working on it.

```

```python
sns.heatmap(fl2)
```

```python
sns.heatmap(fl2, annot=True)
```

```python
sns.heatmap(fl2, annot=True, cmap='rainbow')
```

```python
fl.pivot_table(values='passengers', index='month', columns='year')
```

```python
tip3 = sns.load_dataset('tips')
tip3
```

```python
tip3_p = tip3.pivot_table(values='total_bill', index='day', columns='time')
tip3_p
```

```python
sns.heatmap(tip3_p, annot=True)
```

```python
sns.heatmap(tip3_p, annot=True, cmap='coolwarm', linecolor='white' )
```

```python
sns.heatmap(tip3_p, annot=True, cmap='coolwarm', linecolor='white', linewidths=1)
```
