```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Lmplot -
  # we can't workwith string columns in Regression plots
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, palette='rainbow', hue='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, palette='rainbow', hue='sex', markers=['o','v'], scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', row='time')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', hue='time', markers=['o','v'], scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', aspect=0.6, height=5)
```
