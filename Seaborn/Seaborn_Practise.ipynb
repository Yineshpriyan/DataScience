```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.set_style('whitegrid')
```

```python
g = sns.JointGrid(x='fare', y='age', data=titanic)
g.plot(plt.scatter, sns.histplot)
```

```python
sns.histplot(x='fare', data=titanic, palette='coolwarm')
```

```python
sns.boxplot(x='class', y='age', data= titanic, palette='rainbow')
```

```python
sns.swarmplot(x='class', y='age', data=titanic, palette='rainbow')
```

```python
sns.countplot(x='sex', data=titanic, palette='rainbow')
```

```python
titanic_p = titanic.pivot_table(values='fare', index='sex', columns='age')
titanic_p
#sns.heatmap(titanic.corr(), palette='coolwarm')
```

```python
sns.heatmap(titanic_p, cmap='coolwarm')
```

```python
fg = sns.FacetGrid(data=titanic, col='sex')
fg.map(sns.histplot,'age')
```
