Matplotlib is used for Data Visualization using Graphs or plots

Types of Plots
-------------
- Line Plot
- Bar Plot
- Scatter Plot
- Pie Plot
- Area Plot
- Histogram Plot

```python
import numpy as np
import matplotlib.pyplot as plt
```

Line Plot

```python
# Initialize Data
x = np.array([1,2,3,4,5])
y = np.array([12,25,40,8,30])

# Plotting
plt.plot(x,y)

# Adding title
plt.title('Performance Report')

# Axis Label
plt.xlabel("Index No")
plt.ylabel('Marks')

# Legend
plt.legend(['Line'])

# Show
plt.show()
```

```python
# Initialize Data
x = np.array([100,101,102,103,104])
y= np.array([80,40,65,91,53])

# Plot the Data
plt.plot(x,y)

# Add Title
plt.title('Line Chart')

# Add Axis Label
plt.xlabel('Student Index No')
plt.ylabel('Marks')

# Legend
plt.legend(['Line'])

# Show
plt.show()
```

```python
# linestyle = 'dotted' / 'dashed' /
x = np.array([6,15,3,20,66])
plt.plot(x, linestyle='dashed')
plt.show()
```

Bar Chart

```python
# Initialize
x = np.array([100,101,102,103,104])
y= np.array([80,40,65,91,53])

# Plot
plt.bar(x,y)

# Add Title
plt.title('Bar Chart')

# Add Axis label
plt.xlabel('Student Index No')
plt.ylabel('Student Marks')

# Legend
plt.legend(['bar'])

# Show
plt.show()
```

Scatter Plot

```python
# Initialize
x = np.array([100,101,102,103,104])
y = np.array([80,40,65,91,53])

# Plot
plt.scatter(x,y)

# Title
plt.title('Scatter Chart')

# Axis Label
plt.xlabel('Student Index No')
plt.ylabel('Student Marks')

plt.legend(['Scatter'])

# Show
plt.show()
```

```python
# Initialize
x = np.array([1,2,3,4,5,6,7])
y = np.array([1,2,3,4,5,6,7])

# Pot
plt.hist(x,y)

# Title
plt.title('Histogram Chart')

# Axis Label
plt.xlabel('Studnet Index No')
plt.ylabel('Student Marks')

# Legend
plt.legend(['Histogram'])

# Show
plt.show()
```

```python
# Initialize
x = np.array([1,2,3,4,5]) # OR x = [1,2,3,4,5]
e = (0,0.1,0,0,0)

# Plot
plt.pie(x, explode = e)

# Title
plt.title('Pie Chart')

# Legend
plt.legend(['Pie'])

# Show
plt.show()
```

3D Plot

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]
z = [1,8,27,64,125]

#fig = plt.figure()
ax = plt.axes(projection = '3d')
ax.plot(z,y,x)
```
