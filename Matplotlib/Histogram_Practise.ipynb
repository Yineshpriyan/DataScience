```python
df = np.random.normal(100,20,250)
df
```

#Simple histogram with custom bins

```python
fig = plt.figure(figsize=(6,3))
plt.hist(df)
```

#Simple histogram with custom ranges instead of bins count

```python
fig = plt.figure(figsize=(6,3))
plt.hist(df, bins=4)
```

```python
fig = plt.figure(figsize=(6,3))
plt.hist(df, bins=[10,30,40,50,55,80])
```

#Load Dataset

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Matplotlib/Ecommerce Customers.csv')
df
```

#customize

```python
fig = plt.figure(figsize=(6,3))
plt.xlabel('X')
plt.ylabel('Y')
plt.hist(df['Avg. Session Length'], bins=5, edgecolor='black', facecolor='green')
```

```python
fig = plt.figure(figsize=(6,3))
plt.hist(df['Avg. Session Length'], bins=5, edgecolor='black', facecolor='green', orientation='horizontal')
```

```python
fig = plt.figure(figsize=(6,3))
plt.hist(df['Avg. Session Length'], bins=5, edgecolor='black', facecolor='green', orientation='horizontal', histtype = 'step')
```

#Density

```python
fig = plt.figure(figsize=(6,3))
plt.hist(df['Avg. Session Length'], bins=5, edgecolor='black', facecolor='green', orientation='vertical', density='True')
```

#Cummulative - Frequency Count

```python
fig = plt.figure(figsize=(6,3))
plt.hist(df['Avg. Session Length'], bins=5, edgecolor='black', facecolor='green', orientation='vertical', cumulative='T')
```

#cummulative density

```python
fig = plt.figure(figsize=(6,3))
plt.hist(df['Avg. Session Length'], bins=5, edgecolor='black', facecolor='green', orientation='vertical', density='T', cumulative='T')
```
