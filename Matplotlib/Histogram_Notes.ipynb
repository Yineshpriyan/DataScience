Histogram is graph shows count, density (frequency distribution) for certain ranges of data
It is looks like Bar chart but Histogram graph works fro intervals / range not exact values+
height of the bar represents cunt/ frequesncy distribution for that range.

Histograms are approximate representations of the distribution of numeric data. Later it can be converted into actual distributions using kernal density estimation.


#Import

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
weights = np.random.normal(100,20,250)
print(weights)
```

#Simple histogram deafault bins (State Base Approach)

```python
fig = plt.figure(figsize=(6,3))
plt.hist(weights)
```

#Simple histogram deafault bins (Object Approach)

```python
fig,ax = plt.subplots()
ax.hist(weights)
```

#Simple histogram with custom bins

```python
fig = plt.figure(figsize=(6,3))
plt.hist(weights,bins=5)
```

#Simple histogram with custom ranges instead of bins count

```python
figure = plt.figure(figsize=(6,3))
plt.hist(weights, bins=[10,50,70,80,90,100,110,200])
```

#Load Dataset

```python
customer_data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Matplotlib/Ecommerce Customers.csv')
customer_data
```

#customize

```python
figure = plt.figure(figsize=(6,3))
plt.xlabel('Avg Session Length')
plt.ylabel('Avg. Session Frequency Count')
plt.hist(customer_data['Avg. Session Length'])
```

```python
fig = plt.figure(figsize=(6,3))
plt.xlabel('Avg Session Length')
plt.ylabel('Avg. Session Frequency Count')
plt.hist(customer_data['Avg. Session Length'],edgecolor='black')
```

```python
fig = plt.figure(figsize=(6,3))
plt.xlabel('Avg Session Length')
plt.ylabel('Avg. Session Frequency Count')
plt.hist(customer_data['Avg. Session Length'], edgecolor='black', facecolor='yellow', bins=20)
```

```python
fig = plt.figure(figsize=(6,3))
plt.xlabel('Avg. Session Frequency Count')
plt.ylabel('Avg Session Length')
plt.hist(customer_data['Avg. Session Length'], edgecolor='black', facecolor='red', orientation='horizontal')
```

```python
fig = plt.figure(figsize=(6,3))
plt.xlabel('Avg. Session Frequency Count')
plt.ylabel('Avg Session Length')
plt.hist(customer_data['Avg. Session Length'], edgecolor='black', facecolor='yellow', orientation='horizontal', histtype='step')
```

```python
fig = plt.figure(figsize=(6,3))
plt.xlabel('Avg. Session Frequency Count')
plt.ylabel('Avg Session Length')
plt.hist(customer_data['Avg. Session Length'], edgecolor='black', facecolor='yellow', orientation='horizontal', histtype='stepfilled')
```

```python
fig = plt.figure(figsize=(6,3))
plt.xlabel('Avg. Session Frequency Count')
plt.ylabel('Avg Session Length')
plt.hist(customer_data['Avg. Session Length'], edgecolor='black', facecolor='yellow', orientation='horizontal', histtype='bar')
```

```python
fig = plt.figure(figsize=(6,3))
plt.xlabel('Avg. Session Frequency Count')
plt.ylabel('Avg Session Length')
plt.hist(customer_data['Avg. Session Length'], edgecolor='black', facecolor='yellow', orientation='horizontal', histtype='barstacked')
```

#Density

```python
fig = plt.figure(figsize=(6,3))
plt.hist(customer_data['Avg. Session Length'], density=True)
```

#Cummulative - Frequency Count

```python
fig = plt.figure(figsize=(6,3))
plt.hist(customer_data['Avg. Session Length'], cumulative=True)
```

#cummulative density

```python
fig = plt.figure(figsize=(6,3))
plt.hist(customer_data['Avg. Session Length'], cumulative = True, density = True)
```
