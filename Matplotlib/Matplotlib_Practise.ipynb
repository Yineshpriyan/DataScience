```python
import matplotlib.pyplot as plt
import numpy as np
```

Matplotlib is a python library to plot data for the use of data visualization.

Types of Graph.
Line
Bar
Scatter
Pie
Histogram
3D

```python
# Initialize
x = np.array(['Ram','Raj','Shyam','Sita','Mala'])
y = np.array([45,78,66,34,85])

# Plot
plt.plot(x,y, linestyle='dashed')

# Title
plt.title('Line Chart')

# Axis Label
plt.xlabel('Student Name')
plt.ylabel('Marks')

# Legend
plt.legend(['Line'])

# Show
plt.show()
```

```python
# Initialize
x = ['Ram','Raj','Shyam','Sita','Mala']
y = [45,78,66,34,85]

# Plot
plt.bar(x,y)

# title
plt.title('Bar Chart')

# Axis
plt.xlabel('Student Name')
plt.ylabel('Marks')

# Legend
plt.legend(['Bar'])

# SHow
plt.show()

```

```python
# Initialize
x = np.array(['Ram','Raj','Shyam','Sita','Mala'])
y = np.array([45,78,66,34,85])

# Plot
plt.scatter(x,y)

# Title
plt.title('Scatter Chart')

# Axis
plt.xlabel('Student Name')
plt.ylabel('Marks')

# Legend
plt.legend(['Scatter'])

# Show
plt.show()
```

```python
# Initialize
x = [45,12,8,20,3,17]
e = [0,0.1,0,0,0.1,0]

# Plot
plt.pie(x, explode = e)

# Title
plt.title('Pie Chart')

# Axis


# Legend
plt.legend(['Pie'])

# Show
plt.show()
```

```python
# Initialize
x = [5,10,15,20]
y = [10,20,30,40]

# Plot
plt.hist(x,y)

# Title
plt.title('Histogram Chart')

# Axis
plt.xlabel('Name')
plt.ylabel('Marks')

# Legend
plt.legend(['Hist'])

# Show
plt.show()
```
