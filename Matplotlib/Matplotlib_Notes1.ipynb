Line Chart

```python
import matplotlib.pyplot as plt
import numpy as np

points = np.array([5,7,18,25,9,5])
plt.plot(points)
plt.show()
```

```python
# linestyle = 'dotted' / 'dashed' /
x = np.array([6,15,3,20,66])
plt.plot(x, linestyle='dashed')
plt.show()
```

Bar Chart

```python
x = np.array(['A','B','C','D'])
y = np.array([10,22,8,30])
plt.bar(x,y)
plt.show()
```

Scatter Chart

```python
x1 = np.array([20,15,11,7,8,10])
y1 = np.array([2,7,15,8,10,12])

plt.scatter(x1,y1)
plt.show()
```

Pie Chart

```python
p = np.array([30,20,25,20])
plt.pie(p)
plt.show()
```
