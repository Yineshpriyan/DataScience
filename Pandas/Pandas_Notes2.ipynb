60 Pandas with Numpy

```python
l1 = [
    {'Brand':'Dell' , 'Price':80000},
    {'Brand': 'Lenova' , 'Price':75000},
    {'Brand': 'HP', 'Price': 90000}
]

df1 = pd.DataFrame(l1)
print(df1)
print()
print(len(df1))
print()

for i in range(len(df1)):
  for col in df1.columns:
    value = df1.at[i,col]
    print(f"{i} value: {value}")



'''
for i in range(len(df1)):
    for col in df1.columns:
        value = df1.at[i, col]  # Access cell value
        print(f"Row {i}, Column '{col}': {value}")
'''


```

69 Add, Rename and Delete Columns in DataFrame


```python
s1 = pd.Series(['Ravi','Kamal','Vikram','Selva'])
df1 = pd.DataFrame(s1)
print(df1)
print(df1.columns)
print()
df1.columns = ['Name']
print(df1)
print(df1.columns)
print()
df1['Marks'] = 50
print(df1)
print()
df1['Average'] = [35,45,55,65]
print(df1)
print()
df1['tot'] = df1['Marks'] + df1['Average']
print(df1)
print()
df1['tot'] = 'Total'
print(df1)
print()

del df1['Average']
print(df1)

df1.pop('Name')
print(df1)


```

70 Deleting column & Row using Drop


```python
s1 = pd.Series(['Ravi','Kamal','Vikram','Selva'])
df1 = pd.DataFrame(s1)
df1.columns = ['Name']
df1['Marks'] = 50
df1['Average'] = ['40','50','60','70']
print(df1)
print()

df2 = df1.drop('Average',axis=1)
print(df2)
print()

df3 = df1.drop(index=[2,3],axis=0)
print(df3)
print()
```

71 Boolean Indexing in DataFrame


```python
dic= {
    'Name': ['Raj','Kamal','Ajith','Vijay'],
    'Age': [70,60,50,40]
}
df1 = pd.DataFrame(dic)
print(df1)
print()

df2 = pd.DataFrame(dic, index=[True,False,True,False])
print(df2)
print()

print(df2.loc[True])
print()

print(df2.loc[False])
print()
```

72 Concatenating DataFrames


```python
dic1 = {
    'Name': ['Rajini', 'Kamal', 'Ajith', 'Vijay'],
    'Age': [66,65,42,40]
}
dic2 = {
    'Name': ['Dhanush', 'Simbu', 'Karthi', 'Jeyam Ravi'],
    'Age': [38,38,37,36]
}

df1 = pd.DataFrame(dic1)
df2 = pd.DataFrame(dic2)
df3 = pd.concat([df1,df2])
print(df3)
print()
df4 = pd.concat([df1,df2])
print(df3)
print()
```
