```python
import pandas as pd
```

```python
# Series - Like a column in a table
# 1-D array

a = [1,7,2]
var = pd.Series(a)
print(a)
print(var)
print()

b= {"day1":'Monday', "day2":'Tuesday', "day3": 3, "day4": 'Friday'}
var2 = pd.Series(b)
print(b)
print(var2)

```

```python
# DataFrame - 2D Array / Table with rows and columns

data = {
    "Month": ['Jan','Feb','Mar','Apr','May','June'],
    "Day": ['Mon', 'Tue', 'Wed', 'Thur','Fri','Sat']
}

df = pd.DataFrame(data)
print(data)
print(df)
print()
```

```python
# Loacte Row

r0= df.loc[0]
print(r0)
print()

print(df.loc[2])
print()

print(df.loc[[0,1,2]])
print()
```
